#  Outlier Treatment (capping & flooring)

### Outlier Treatment for n_hot_rooms as shown in the plotted graph for some values above 99th percentile

```python
np.percentile(df.n_hot_rooms,[99])
```

```python
np.percentile(df.n_hot_rooms,[99])[0]
```

```python
uv = np.percentile(df.n_hot_rooms,[99])[0]
```

```python
df[(df.n_hot_rooms>uv)]
```

```python
df.n_hot_rooms[(df.n_hot_rooms>3*uv)]=3*uv
```

### Outlier Treatment for rainfall as shown in the plotted graph for some values below 1th percentile

```python
np.percentile(df.rainfall,[1])[0]
```

```python
lv = np.percentile(df.rainfall,[1])[0]
```

```python
df[(df.rainfall< lv)]
```

```python
df.rainfall[(df.rainfall <0.3*lv)] = 0.3*lv
```

```python
sns.jointplot(x="crime_rate", y="price", data=df)
```

```python
df.describe()
```

```python
df.info()
```

### Missing values in n_hos_beds can be taken as the mean of all hos beds

```python
df.n_hos_beds = df.n_hos_beds.fillna(df.n_hos_beds.mean())
```

```python
df.info()
```

### Skewness or outliers in crime rate treatment

```python
sns.jointplot(x="crime_rate", y="price",data=df)
```

```python
df.crime_rate = np.log(1+df.crime_rate)
```

```python
sns.jointplot(x="crime_rate", y="price",data=df)
```

### Multiple Linear Regression Model

#### Mutliple linear regression model through statsmodels.api library

```python
import statsmodels.api as sn
```

```python
X_multi = df.drop("price",axis=1)
```

```python
X_multi.head()
```

```python
y_multi = df['price']
```

```python
y_multi.head()
```

```python
X_multi_cons = sn.add_constant(X_multi)
```

```python
X_multi_cons.head()
```

```python
lm_multi = sn.OLS(y_multi, X_multi_cons).fit()
```

```python
lm_multi.summary()
```

#### Mutliple linear regression model through sklearn library

```python
from sklearn.linear_model import LinearRegression
```

```python
lm3 = LinearRegression()
```

```python
lm3.fit(X_multi, y_multi)
```

```python
print(lm3.intercept_, lm3.coef_)
```
